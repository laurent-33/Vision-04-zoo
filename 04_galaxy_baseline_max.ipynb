# Autres baselines

```python
def convert_baseline(csv_baseline_file):
    Y_trainfull = pd.read_csv(csv_baseline_file, index_col=0)
    Y_trainfull = Y_trainfull.iloc[:, 0:3]
    Y_trainfull['max'] = Y_trainfull[["Class1.1", "Class1.2", "Class1.3"]].idxmax(axis=1)
    
    Y_trainfull = pd.get_dummies(Y_trainfull)
    Y_trainfull = Y_trainfull.iloc[:, 3:6]

    return Y_trainfull
```

```python
convert_baseline('./csv/central_pixel_benchmark.csv')
```
